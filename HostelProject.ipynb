### Data encoding and feature generation

```python
def dataModification(df):
    mapping = {'A': 0, 'P': 1}
    for i in range(1,31):
        df[str(i)]=df[str(i)].map(mapping)
    dates=[]
    for i in range(1,31):
        dates.append(str(i))
    df['TotalPresent'] = df[dates].sum(axis=1)
    messWiseSortedDf = df.sort_values(by='MessType')
    return messWiseSortedDf
```

### No. of students enrolled in the mess

```python
def messWiseStudentCount(df):
    grouped_df = df.groupby('MessType')
    group_count=grouped_df.size()
    return (group_count, grouped_df)
```

### Total meals in each mess


```python
def totalMealCount(grouped_df):
    for group_name, group_data in grouped_df:
        print(f"Mess Caterer: {group_name}")
        print('Total meals')
        print(sum(group_data['TotalPresent']))
        print()
```

### Attendance report of each student mess wise

```python
def generateAttendanceReport(grouped_df):
    days=int(input('Data required till which date: '))
    columns=['Name','RegistrationNo','Block','MessType']
    for i in range(1,days+1):
        columns.append(str(i))
    group_dfs = []
    group_names=[]
    for group_name, group_data in grouped_df:
        group_dfs.append(group_data)
        group_names.append(group_name)
        
    for i, group_df in enumerate(group_dfs):
        print(group_names[i])
        print(group_df[columns])
        print()
        print(sum(group_df['TotalPresent']))
        print()
    
    
```
