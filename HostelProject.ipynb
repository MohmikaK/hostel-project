```python
import pandas as pd
import random
import string
```

### Data generation

```python
num_names = 1000

names = [''.join(random.choices(string.ascii_letters, k=random.randint(5, 10))) for _ in range(num_names)]

df = pd.DataFrame({'Name': names})
yy=  [''.join(random.choices(['19','20','21','22'], k=1)) for _ in range(num_names)]
bra=  [''.join(random.choices(['BCB','BCE','BCT','BDS','BKT','BBS'], k=1)) for _ in range(num_names)]
regn=[]
for i in range(num_names):
    rand= random.randint(1, 4000)
    numb = str(rand).zfill(4)
    s=yy[i]+bra[i]+numb
    regn.append(s)
df['RegistrationNo']=regn
```

```python
block=  [''.join(random.choices(['A','B','C','D','E','F','G','H'], k=1)) for _ in range(num_names)]
df['Block']=block
mess=  [''.join(random.choices(['Sakthi Veg','Sakthi Non-Veg', 'Darling-special', 'Neelkesh Veg', 'Neelkesh Special'], k=1)) for _ in range(num_names)]
df['MessType']=mess
```

```python
for i in range(1,31):
    atten=  [''.join(random.choices(['P','A'], [0.8,0.2], k=1)) for _ in range(num_names)]
    date=str(i)
    df[date]=atten
```

```python
df
```

### Data encoding and feature generation

```python
mapping = {'A': 0, 'P': 1}
for i in range(1,31):
    df[str(i)]=df[str(i)].map(mapping)
```

```python
dates=[]
for i in range(1,31):
    dates.append(str(i))
df['TotalPresent'] = df[dates].sum(axis=1)
```

```python
df
```

```python
sorted_df = df.sort_values(by='MessType')
```

### No. of students enrolled in the mess

```python
grouped_df = df.groupby('MessType')
group_count=grouped_df.size()
print(group_count)
```

### Total meals in each mess


```python
for group_name, group_data in grouped_df:
    print(f"Mess Caterer: {group_name}")
    print('Total meals')
    print(sum(group_data['TotalPresent']))
    print()
```
